```python
import matplotlib.pyplot as plt
plt.hist(df['price'], bins=50, color='blue', edgecolor='black')
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(x=df['airline'] , y=df['price'])
plt.xticks(rotation = 45)
plt.xlabel('Airlines')
plt.ylabel('Price')
plt.title('Boxplot diagram')
plt.show()
```

5.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(y=df['price'], color='lightblue')
plt.ylabel('Flight Price')
plt.title('Boxplot of Flight Prices')
plt.show()
```

```python
import numpy as np

Q1 = df['price'].quantile(0.25)  # First quartile (25th percentile)
Q3 = df['price'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile range

lower_bound = Q1 - 1.5 * IQR  # Lower threshold
upper_bound = Q3 + 1.5 * IQR  # Upper threshold

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print(f"Number of Outliers: {outliers.shape[0]}")
```
